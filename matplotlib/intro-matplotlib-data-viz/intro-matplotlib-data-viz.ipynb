## Two Approaches for creating Plots

### 1. Functional Approach:

```python
x = np.linspace(0, 10, 20) # Generate 20 datapoints between 0 and 10
y = x**2                   # 
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)

plt.title("Our first plot")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
# plt.subplt(nrows, ncols, plot_number)
plt.subplot(1, 2, 1)
plt.plot(x, y,"red")

plt.subplot(1, 2, 2)
plt.plot(y, x, "green")
```

### 2. Object oriented Interface:

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.2, 0.8, 0.9])
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.2, 0.8, 0.9])

ax.plot(x, y, "purple")
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.2, 0.8, 0.9])

ax.plot(x, y, "purple")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Our First Plot using Object Oriented Approach")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)
```

```python
# Empty canvas of 3 by 3 subplots
fig, axes = plt.subplots(nrows=3, ncols=3)
```

```python
# Empty canvas of 3 by 3 subplots
fig, axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout()
```

```python
# Empty canvas of 3 by 3 subplots
fig, axes = plt.subplots(nrows=3, ncols=3)

axes[0, 1].plot(x, y)
axes[1, 2].plot(y, x)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 2), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

## Legends

```python
fig = plt.figure(figsize=(8, 6), dpi=60)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2)
ax.plot(x, x**3, "red")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=60)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label="X Square Plot")
ax.plot(x, x**3, "red", label="X Cube Plot")

ax.legend()
```

## Plot Appearance

```python
fig = plt.figure(figsize=(8, 6), dpi=60)
ax = fig.add_axes([0, 0, 1, 1])

# Change the line thickness to 3, line style to dashes nad markout the datapoints
ax.plot(x, y, color="purple", linewidth=3, linestyle="--", marker="o", markersize=8)
```

## Plot range

```python
fig = plt.figure(figsize=(8, 6), dpi=60)
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color="purple", lw=3, ls="--")
ax.set_xlim([0, 1])  # [0, 1] signifies the lower bound and upper bound of x axis
ax.set_ylim([0, 5])  # [0, 5] signifies the lower bound and upper bound of y axis
```

## Sample Application of Visualization

```python
my_data = pd.read_csv("./nations.csv")
```

```python
my_data
```

```python
# Replace missing values with the median value of that column

my_data["gdp_percap"].fillna(my_data["gdp_percap"].median(), inplace=True)
```

```python
my_data.head(5)
```

```python
my_data.groupby(['country']).mean()
```

```python
avg_gdp_percap = my_data.groupby(["country"]).mean()["gdp_percap"]

avg_gdp_percap
```

```python
top_five_countries = avg_gdp_percap.sort_values(ascending=False).head()
```

```python
top_five_countries
```

```python
united_arab_emirates = my_data[my_data["country"] == "United Arab Emirates"]

united_arab_emirates
```

```python
plt.plot(united_arab_emirates["year"], united_arab_emirates["gdp_percap"])

plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("GDP per Capita United Arab Emirates")
```

```python
united_arab_emirates.plot.bar(x="year", y="gdp_percap")
```

```python
plt.subplot(3, 1, 1)

plt.title("GDP Per Capita")
plt.plot(united_arab_emirates["year"], united_arab_emirates["gdp_percap"])

plt.subplot(3, 1, 2)

plt.title("GDP in Billions")
plt.plot(united_arab_emirates["year"], (united_arab_emirates["population"] * united_arab_emirates["gdp_percap"] / 10**9))

plt.subplot(3, 1, 3)

plt.title("Population in millions")

plt.plot(united_arab_emirates["year"], united_arab_emirates["population"] / 10**6)

plt.tight_layout()
```

```python
plt.subplot(3, 1, 1)

plt.title("GDP Per Capita")
plt.bar(united_arab_emirates["year"], united_arab_emirates["gdp_percap"], color="r")

plt.subplot(3, 1, 2)

plt.title("GDP in Billions")
plt.bar(united_arab_emirates["year"], (united_arab_emirates["population"] * united_arab_emirates["gdp_percap"] / 10**9), color="g")

plt.subplot(3, 1, 3)

plt.title("Population in millions")

plt.bar(united_arab_emirates["year"], united_arab_emirates["population"] / 10**6, color="b")

plt.tight_layout()
```

```python
plt.plot(united_arab_emirates["year"], (united_arab_emirates["population"] / united_arab_emirates["population"].iloc[0] * 100))

united_arab_emirates_gdp = united_arab_emirates["population"] * united_arab_emirates["gdp_percap"]

plt.plot(united_arab_emirates["year"], united_arab_emirates_gdp / united_arab_emirates_gdp.iloc[0] / 100)

plt.plot(united_arab_emirates["year"], united_arab_emirates["gdp_percap"] / united_arab_emirates["gdp_percap"].iloc[0] / 100)

plt.title("GDP and Population Growth in United Arab Emirates (first year = 100)")

plt.legend(["Population", "GDP", "GDP Per Capita"], loc=4)
```

```python
plt.bar(united_arab_emirates["year"], (united_arab_emirates["population"] / united_arab_emirates["population"].iloc[0] * 100), color="r")

united_arab_emirates_gdp = united_arab_emirates["population"] * united_arab_emirates["gdp_percap"]

plt.bar(united_arab_emirates["year"], united_arab_emirates_gdp / united_arab_emirates_gdp.iloc[0] / 100, color="g")

plt.bar(united_arab_emirates["year"], united_arab_emirates["gdp_percap"] / united_arab_emirates["gdp_percap"].iloc[0] / 100, color="b")

plt.title("GDP and Population Growth in United Arab Emirates (first year = 100)")

plt.legend(["Population", "GDP", "GDP Per Capita"], loc=4)
```

```python
# Comparing Qatar with United Arab Emirates
qt = my_data[my_data["country"] == "Qatar"]

plt.bar(qt["year"], qt["gdp_percap"])
plt.bar(united_arab_emirates["year"], united_arab_emirates["gdp_percap"])
plt.legend(["Qatar", "United Arab Emirates"])
```
