# 5 Advanced Visualisation for Exploratory data analysis (EDA)

Source code for Medium's article ["5 Advanced Visualisation for Exploratory data analysis (EDA)"](https://towardsdatascience.com/5-advanced-visualisation-for-exploratory-data-analysis-eda-c8eafeb0b8cb), written by [Kaushik Choudhury]().

## Step 1

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2

```python
bird = sns.load_dataset("penguins")

bird
```

```python
bird.info()
```

```python
g = sns.FacetGrid(bird, col="island", hue="species")
g.map(plt.scatter, "flipper_length_mm", "body_mass_g", alpha=.6)
g.add_legend()
plt.show()
```

### Step 3

```python
sns.stripplot(x="island", y="body_mass_g", hue="species", data=bird,
              palette="Set1")
plt.show()
```

```python
sns.swarmplot(x="island", y="body_mass_g", hue="species", data=bird,
              palette="Set1", dodge=True)

plt.show()
```

## Step 4

```python
sns.lmplot(x="body_mass_g", y="flipper_length_mm", hue="sex", col="island",
           markers=["o", "x"], palette="Set1", data=bird)

plt.show()           
```

```python
sns.lmplot(x="body_mass_g", y="flipper_length_mm", hue="sex", col="island",
          row="sex", order=2, markers=["o", "x"], palette="Set1", data=bird)

plt.show()       
```

## Step 5

```python
sns.set_palette("gist_rainbow_r")
sns.jointplot(x="body_mass_g", y="flipper_length_mm", kind="hex", data=bird)

plt.show()
```

```python
g = sns.jointplot(x="body_mass_g", y="flipper_length_mm", data=bird,
                  kind="hex", color="c")

g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.set_axis_labels("Body Mass (in gram)", "Flipper Length (in mm)")

plt.show()
```

## Step 6

```python
sns.pairplot(bird, hue="island")

plt.show()
```
